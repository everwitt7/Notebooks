```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a random distribution of data
points = 100
x = np.random.normal(0, 1, points)
y = x**3 + 25*np.random.normal(0, 0.1, points)

# kernel smoother
def kernel_smoother(x, y, h):
    # create a grid of x values
    x_grid = np.linspace(min(x), max(x), 100)
    # create a list to store the smoothed values
    y_smooth = []
    # loop through each value in the grid
    for x_val in x_grid:
        # calculate the weights
        weights = np.exp(-(x - x_val)**2/(2*h**2))
        # calculate the smoothed value
        y_smooth.append(np.sum(weights*y)/np.sum(weights))
    return x_grid, y_smooth

xx, yy = kernel_smoother(x, y, 0.1)
```

```python
plt.scatter(x, y)
plt.plot(xx, yy)
```
