```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
# generate random numbers from a exponential distribution
shape, samples = 5, 1000
data = np.random.lognormal(mean=0.0, sigma=1.0, size=samples)
```

```python
# create subplots
f, axs = plt.subplots(1, 2, figsize=(18,8))

sns.histplot(data=data, kde=True, ax=axs[0])
axs[0].set_title("Original Distribution")
axs[0].set_xlabel("Value")
axs[0].set_ylabel("Count")

sns.histplot(data=np.log(data), kde=True, ax=axs[1])
axs[1].set_title("Log Transformation Distribution")
axs[1].set_xlabel("Value")
axs[1].set_ylabel("Count")
```

```python
x = np.linspace(0.1, 100, 200)
data = {
    "ln(x)": np.log(x),
    "log2(x)": np.log2(x),
    "log7(x)": np.log(x)/np.log(7),
    "log10(x)": np.log10(x),
}
df = pd.DataFrame(data, index=x)
```

```python
# generate data
x = np.linspace(0.1, 5, 200)
f_x = np.e**x
g_x = np.log(f_x) # transformed
p_x = f_x - 100 # shifted
q_x = p_x / 50 # scaled

# plot f(x) and g(f(x))
f, axs = plt.subplots(2, 2, figsize=(16,8))
sns.lineplot(x=x, y=f_x, ax=axs[0][0])
sns.lineplot(x=x, y=g_x, ax=axs[0][1])
sns.lineplot(x=x, y=p_x, ax=axs[1][0])
sns.lineplot(x=x, y=q_x, ax=axs[1][1])

# set graph titles
axs[0][0].set_title("f(x) = e^(x)")
axs[0][1].set_title("g(f(x)) = ln(e^x)")
axs[1][0].set_title("p(f(x)) = e^(x) - 100")
axs[1][1].set_title("q(f(x)) = e^x / 50")
```
